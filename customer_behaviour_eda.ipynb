```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
labels = ['young_adult','adult','middle_aged','senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df.columns
```

```python
frequency_mapping = {'fortnightly': 14,
                     'weekly': 7,
                     'monthly': 30,
                     'quarterly': 90,
                     'bi-weekly': 14,
                     'annually':365,
                     'every 3 months': 90} # Corrected key 'every_3_months' to 'every 3 months'

df['purchase_frequency_days']= df['frequency_of_purchases'].str.lower().map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
df.isnull().sum()
```

```python
unmapped_frequencies = df[df['purchase_frequency_days'].isnull()]['frequency_of_purchases'].unique()
print("Unmapped frequencies:", unmapped_frequencies)
```
